```python
"""

This is a K-means Classification module


To use this module
Command:
   >> import k_means as km

1. Create an object
   >> stat = knn.clussterit( matrix of (m examples X n features ) )

2. Feed required data
   >> stat.getcluster ( number of cluster desired , number of iteration )
   This command will output the number of cluster found
   
3. To get the examples of a particular class - 
   >> feature_matrix = stat.class_set( class desired )    --- The class will be labelled as 0,1,2.. and so on....
   
4. To get the label of all examples (in same order as they were fed in) - 
   >> all_labels = stat.labels

"""
```

```python
import scipy.io as sio
import k_means as km
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing data

k = sio.loadmat('ex7data1.mat')
X = np.array(k['X'])
```

```python
# creating object and setting parameters -  ( number of cluster desired , number of iteration )

stat = km.clusterit(X)
stat.getcluster(3,100)
```

```python
# finding out class 0 examples


print stat.class_set(0)
```

```python
# AS the data used has only 2 features , we can here visualise data

plt.scatter(X[:,0],X[:,1],c=stat.labels)
plt.show()
```
